```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread("pic.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# hist: 256x1 array
hist = cv.calcHist([img],[0],None,[256],[0,256])

plt.subplot(121),plt.imshow(img),plt.title("hist")
plt.xticks([]), plt.yticks([])

plt.subplot(122)

#grayscale histogram plot
# plt.hist(img.ravel(),256,[0,256]),plt.title("hist")
# plt.xticks([]), plt.yticks([])

#bgr histogram plot
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])

plt.show()
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('wiki.jpg',0)

eq = cv.equalizeHist(img)
res = np.hstack((img,eq)) #stack them sidebyside
res = cv.cvtColor(res, cv.COLOR_GRAY2RGB)

plt.imshow(res)
plt.show()
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('statue.png',0)

# create a CLAHE object (Arguments are optional).
clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(img)
cl1 = cv.cvtColor(cl1, cv.COLOR_GRAY2RGB)

plt.imshow(cl1)
plt.show()
```

```python
import numpy as np
import cv2 as cv

roi = cv.imread('red.jpg')
hsv = cv.cvtColor(roi,cv.COLOR_BGR2HSV)

target = cv.imread('pic.jpg')
hsvt = cv.cvtColor(target,cv.COLOR_BGR2HSV)

# histogram
roihist = cv.calcHist([hsv],[0,1],None,[180,256],[0,180,0,256])

# normalize histogram
roihist = cv.normalize(roihist,0,255,cv.NORM_MINMAX)
dst = cv.calcBackProject([hsvt],[0,1],roihist,[0,180,0,256],1)

# Now convolute with circular disc
disc = cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
cv.filter2D(dst,-1,disc,dst)

# threshold and binary AND
ret,thresh = cv.threshold(dst,50,255,0)
thresh = cv.merge((thresh,thresh,thresh))
res = cv.bitwise_and(target,thresh)
res = np.vstack((target,thresh,res))
cv.imwrite('res.jpg',res)

```
