```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = np.zeros((512,512,3), np.uint8) # black background

cv.line     (img,(0,0)  ,(511,511),(255,0,0),5)
cv.rectangle(img,(384,0),(510,128),(0,240,0),3)
cv.circle   (img,(447,63),63,(0,0,255), -1)
cv.ellipse  (img,(256,256),(100,50),0,0,180,255,-1)

# yellow corner
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(img,[pts],True,(0,255,255))

# text
font = cv.FONT_HERSHEY_TRIPLEX
cv.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2,cv.LINE_AA)

plt.imshow(img, interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

img: img source

color: for BGR, pass it as a tuple, eg: (255,0,0) for blue. For grayscale, just pass the scalar value

thickness: if -1 is passed for closed figures like circles, it will fill in (default: 1)

lineType: 8-connected, anti-aliased line etc (default: 8-connected) (cv.LINE_AA gives anti-aliased line which looks great for curves)
